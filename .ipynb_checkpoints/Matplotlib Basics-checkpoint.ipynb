# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Methods of creating a plot

```python
#1st method
fig = plt.figure() #Creates a figure
ax = fig.add_subplot() #adds axes
plt.show()
```

```python
#2nd method
x = (1,2,3,4,5)
y = (5,10,15,20,25)
fig = plt.figure() #creates a figure
ax = fig.add_axes([1,1,1,1]) #ads axes
ax.plot(x,y)
plt.show()
```

```python
#3rd method
fig, ax =plt.subplots() # more than 1 plot come out at the same time
ax.plot(x,y);
#figure obj is the big blue square
#axes object, the subplots based from the anatomy 
```

#### Matplotlib example workflow

```python
# 0. import matplotlib
#%matplotlib inline
#import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
#figsize is height and width (dimensions of the plot in width/height)
fig,ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot",
       xlabel="x-axis",
       ylabel='y-axis')

#5. Save &  show the figure
fig.savefig("images/sample-plot.png")
```

#### Making plots with Numpy arrays

##### Line plot

```python
#make 100 elements from 0 to 10 that are evenly spaced out
x = np.linspace(0,10,100)
x[:10]
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

##### Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x))

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

##### Bar plot

```python
#plotting with a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
#adding customization
ax.set(title="Dan's Nut Butter Store",
       ylabel ="Price ($)");

#making a horizontal
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

##### Histogram

```python
#frequency of numerical values/or the graph of the mode
x = np.random.randn(1000)
#np.random.radn is naturally a gaussian distribution
fig,ax=plt.subplots()
ax.hist(x)
```

##### Making subplots

```python
#Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots (nrows=2,
                                            ncols=2,
                                            figsize=(10,5))

#Plot to each data
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Plotting from Pandas DataFrames (Non-Object Oriented Method)
*Used for simple plots

##### Adding columns to car_sales

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
#If this was a string, we'd have to clean it with these functions first
#Remove punctuations
car_sales["Price"] = car_sales["Price"].str.replace(r'[\$,\s\.]', '', regex=True)
car_sales
#Remove decimals
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
#co2_data["population"] = co2_emission_data["population"].str[:-2]
#if this was string, you'd have to cast it into int to use cumsum

#adding ascending sale date column
car_sales["Sale Date"] = pd.date_range("1/1/2023",periods=(len(car_sales)))
car_sales
#adding total sales column
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

##### Plotting updated car_sales

```python
car_sales.plot(x="Sale Date", y="Total Sales")
#have to cast car_sales price int in order to be recognized as int for a scatter plot
car_sales["Price"] = car_sales["Price"].astype(int)
#scatter
car_sales.plot(x="Odometer (KM)", y="Price", kind ="scatter")
#horizontal  bar
car_sales.plot(x="Make", y="Price",kind="barh")
#histogram
car_sales.plot(x="Make",y = "Odometer (KM)",kind="hist")
```

### Plotting from Pandas DataFrames Object Oriented Method
*Used for plotting more advanced

```python
hd_data = pd.read_csv("heart-disease.csv")
hd_data.head(25)
```

```python
#Scatter plot of age
over_50 = hd_data[hd_data["age"]>50]
over_50
#Pyplot method
over_50.plot(kind='scatter',
               x='age',
               y='chol',
               c='target',
               figsize=(10,6))

# OO Method mixed with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

#sets limitation for ur axis for x
#ax.set_xlim([45,100])
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
# Add a legend
#Grab the values from the scatter, unpack the legend elements and plot it according to the target column values from the table
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
# Subplot of chol, age , thalach
# Creating a single figure w/ 2 axeses
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)

# Add data to ax 0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
# Add a legend to ax0
#Grab the values from the scatter, unpack the legend elements and plot it according to the target column values from the table
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal mean line to ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="thalach");

# Add a horizontal mean line to ax0
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--")

# Add a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold")
```

#### Customizing with the set() method

```python
# Set plot style 
plt.style.use('seaborn-v0_8-whitegrid')
# OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")
# cmap link: https://matplotlib.org/stable/users/explain/colors/colormaps.html
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")
# Add a legend
#Grab the values from the scatter, unpack the legend elements and plot it according to the target column values from the table
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

##### Recreation of two subplots with full customization

###### This plot shows heart disease analysis correlation

```python
# Subplot of chol, age , thalach
# Creating a single figure w/ 2 axeses
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex=True)

# Add data to ax 0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
# Change the x axis limits
ax0.set_xlim([50,80])
# Add a legend to ax0
#Grab the values from the scatter, unpack the legend elements and plot it according to the target column values from the table
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal mean line to ax0
ax0.axhline(over_50["chol"].mean(),
          linestyle="--")

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate")
# Change the x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([50,200])
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal mean line to ax0
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--")

# Add a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16, fontweight="bold")

#Right click to save or use this method
# fig.savefig("heart-disease-analysis-plot-saved-with-code.png")
```
