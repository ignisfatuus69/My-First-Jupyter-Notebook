# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Show columns and data types

```python
#Showing all columns
sr_data = pd.read_csv("specialization_ratings.csv")
sr_data.info()
sr_data.head(10)
```

### Showing null values

```python
sr_data.isna().sum()
```

### Showing all distinct values for these specific columns

```python
columns_to_check = ['specialization_name', 'q4_score', 'q5_score', 'q6_score', 'q7_score', 'q8_score', 'q9_score']

unique_values = {}
for column in columns_to_check:
    unique_values[column] = sr_data[column].unique()

# Print or use these unique values as needed
for column, unique in unique_values.items():
    print(f"Unique values for {column}:", unique)
```

### Graphing Overall Satisfaction per Specialization

```python
import matplotlib.pyplot as plt

# Assuming sr_data is your DataFrame
# Group by 'specialization_name' and calculate mean of 'q8_score' for each specialization
specialization_scores = sr_data.groupby('specialization_name')['q8_score'].mean().reset_index()

# Plotting a bar chart for q8_score against specialization_name
plt.figure(figsize=(12, 6))
bars = plt.bar(specialization_scores['specialization_name'], specialization_scores['q8_score'], color='skyblue')
plt.xlabel('Specialization Name')
plt.ylabel('Average q8 Score')
plt.title('Average q8 Score for each Specialization')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better visibility
plt.tight_layout()

# Define different colors for each range
color_4_to_5 = 'green'
color_3_to_4 = 'orange'
color_3_and_below = 'red'

# Filter specializations within the 4-5 average q8 score range and color the corresponding bars
filtered_specializations_4_to_5 = specialization_scores[(specialization_scores['q8_score'] >= 4) & (specialization_scores['q8_score'] <= 5)]
for idx in filtered_specializations_4_to_5.index:
    bars[idx].set_color(color_4_to_5)

# Filter specializations within the 3-4 average q8 score range and color the corresponding bars
filtered_specializations_3_to_4 = specialization_scores[(specialization_scores['q8_score'] >= 3) & (specialization_scores['q8_score'] < 4)]
for idx in filtered_specializations_3_to_4.index:
    bars[idx].set_color(color_3_to_4)

# Filter specializations with an average q8 score of 3 or below and color the corresponding bars
filtered_specializations_3_and_below = specialization_scores[specialization_scores['q8_score'] < 3]
for idx in filtered_specializations_3_and_below.index:
    bars[idx].set_color(color_3_and_below)

# Show the plot
plt.show()

# Print the number of specializations within each filtered range
num_specializations_4_to_5 = len(filtered_specializations_4_to_5)
num_specializations_3_to_4 = len(filtered_specializations_3_to_4)
num_specializations_3_and_below = len(filtered_specializations_3_and_below)

print(f"Number of specializations with average q8 score between 4 and 5: {num_specializations_4_to_5}")
print(f"Number of specializations with average q8 score between 3 and 4: {num_specializations_3_to_4}")
print(f"Number of specializations with average q8 score of 3 and below: {num_specializations_3_and_below}")
```
